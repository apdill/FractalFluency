```python
import numpy as np
from FF.image_processing import thresh, prepare_scalemap, edgedetector
from FF.fractal_generation import midpoint_displacement

from fracstack import portfolio_plot
```

```python
iterations = 11
P = 1
zslice = 0.5

min_size = 16
n_decades = 1.5
max_size = np.ceil(10**(np.log10(min_size)+n_decades))
num_sizes = 50
num_pos = 100

save_dir = r'/home/apd/Projects/FractalFluency/datasets/fractal_portfolio/coastlines'
```

## *Coastline Fractal Portfolio*


Below are some examples of generated coastline fractals using the midpoint displacement algorithm. I calculate the fractal dimension (D0) using the box-counting method and the information dimension (D1) via Shannon entropy. Results aren't always the same when re-executing cells, but should be within neighborhood of the same value. 

Each generated image is 2049x2049 pixels, with analysis spanning 1.5 decades of size from 16 to 506 pixels, which is ~1/4 image size. For each box size, 100 separate offsets are used to determine the measure; for D0, the minimum coverage count from all 100 offsets is used, and for D1, the average entropy of all 100 offsets is used.

---------------------------------------------------

## **D = ~1.1**

```python
target_D0 = 1.1
f_name = f'coastline_example_D0_{target_D0}.png'
D = 1

generated_fractal = None
attempts = 0

while generated_fractal is None and attempts < 100:

    meshmap = midpoint_displacement(iterations, P=P, D=D)
    scalemap = prepare_scalemap(meshmap)
    threshmap, bwratio = thresh(scalemap, iterations, gaussian_blur=True, sigma=4)
    coastline = edgedetector(threshmap).astype(np.uint8)
    generated_fractal = portfolio_plot(coastline, min_size, max_size, num_sizes, num_pos, (21,7), save_dir, f_name, target_D0)
    
    attempts += 1
```
