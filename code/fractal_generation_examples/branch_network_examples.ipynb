```python
import numpy as np
from scipy.interpolate import interp1d

def get_network_params(target_D0, method='linear'):
    """
    Get interpolated network parameters for a given target D0.
    
    Parameters:
    -----------
    target_D0 : float
        Target fractal dimension (should be between 1.1 and 1.9)
    method : str
        Interpolation method ('linear', 'cubic', etc.)
        
    Returns:
    --------
    dict
        Dictionary containing both neuron and network parameters
    """
    # Original data points
    d0_points = np.array([1.1, 1.2, 1.3, 1.4, 1.5, 1.6, 1.7, 1.8, 1.9])
    
    # Parameters that change
    changing_params = {
        'depth': [3, 5, 7, 7, 8, 9, 9, 9, 11],
        'D': [1.0, 1.0, 1.3, 1.6, 1.6, 1.6, 1.7, 1.8, 2.0],
        'mean_branches': [0.75, 0.75, 1.0, 1.0, 1.1, 1.1, 1.2, 1.2, 1.35],
        'num_neurons': [3, 4, 4, 5, 6, 6, 10, 15, 20],
        'total_length': [500, 500, 600, 600, 600, 600, 600, 600, 600]
    }
    
    # Create interpolation functions for each parameter
    interpolators = {
        param: interp1d(d0_points, values, kind=method, bounds_error=False, fill_value='extrapolate')
        for param, values in changing_params.items()
    }
    
    # Get interpolated values
    interpolated_params = {
        param: float(np.round(interpolator(target_D0),2))
        for param, interpolator in interpolators.items()
    }
    
    # Ensure integer values for certain parameters
    integer_params = ['depth', 'num_neurons']
    for param in integer_params:
        interpolated_params[param] = int(interpolated_params[param])
    
    # Add constant parameters
    constant_params = {
        'mean_soma_radius': 1,
        'std_soma_radius': 1,
        'branch_angle': np.pi / 3,
        'weave_type': 'Gauss',
        'randomness': 0.1,
        'curviness': 'Gauss',
        'curviness_magnitude': 1,
        'n_primary_dendrites': 3,
        'initial_thickness': 5,
        'width': 2049,
        'height': 2049,
        'edge_margin': 200
    }
    
    # Split into neuron and network parameters
    neuron_params = {
        'depth': interpolated_params['depth'],
        'mean_soma_radius': constant_params['mean_soma_radius'],
        'std_soma_radius': constant_params['std_soma_radius'],
        'D': interpolated_params['D'],
        'branch_angle': constant_params['branch_angle'],
        'mean_branches': interpolated_params['mean_branches'],
        'weave_type': constant_params['weave_type'],
        'randomness': constant_params['randomness'],
        'curviness': constant_params['curviness'],
        'curviness_magnitude': constant_params['curviness_magnitude'],
        'n_primary_dendrites': constant_params['n_primary_dendrites'],
        'initial_thickness': constant_params['initial_thickness'],
        'total_length': interpolated_params['total_length']
    }
    
    network_params = {
        'width': constant_params['width'],
        'height': constant_params['height'],
        'num_neurons': interpolated_params['num_neurons'],
        'edge_margin': constant_params['edge_margin']
    }
    
    return {'neuron_params': neuron_params, 'network_params': network_params}


test_D0 = 1.45
params = get_network_params(test_D0, method='linear')

d0_range = np.linspace(1.1, 1.9, 100)
changing_params = ['depth', 'D', 'mean_branches', 'num_neurons', 'total_length']

fig, axes = plt.subplots(len(changing_params), 1, figsize=(10, 12))
fig.suptitle('Parameter Interpolation vs Target D0', fontsize=14)

for i, param in enumerate(changing_params):
    # Get interpolated values across the range
    values = [get_network_params(d0)['neuron_params'].get(param) or 
              get_network_params(d0)['network_params'].get(param) 
              for d0 in d0_range]
    
    # Plot original points and interpolation
    axes[i].plot(d0_range, values, '-', label='Interpolated')
    # Plot original points
    original_values = params_data[param]
    axes[i].plot(params_data['target_D0'], original_values, 'o', label='Original')
    
    axes[i].set_ylabel(param)
    axes[i].grid(True)
    axes[i].legend()
    axes[i].margins(x=0.05)

axes[-1].set_xlabel('Target D0')
plt.tight_layout()
plt.show()
```
